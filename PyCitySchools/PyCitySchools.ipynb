```python
# Dependencies and Setup
import pandas as pd
import os
import csv

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# count of schools (len of unique school names)
school_count = len(school_data_complete["school_name"].unique())

# count of student ids
student_count = school_data_complete["Student ID"].count()

# sum of budget for all students
school_budget = 0     
school_budget = school_data["budget"].sum()

# mean of math scores
school_math_mean = school_data_complete["math_score"].mean()

# mean of reading scores
school_rdg_mean = school_data_complete["reading_score"].mean()


# select only passing scores for each subject
pass_math_a = school_data_complete["math_score"]>=70
pass_rdg_a = school_data_complete["reading_score"]>=70

# select only passing scores for each subject
pass_math = school_data_complete[school_data_complete["math_score"]>=70]
pass_rdg = school_data_complete[school_data_complete["reading_score"]>=70]

# count only those passsing scores
pass_math_count = pass_math["Student ID"].count()
pass_rdg_count = pass_rdg["Student ID"].count()

# finding those that pass both reading and math - a functional inner join by passing criteria for both math and reading.
pass_math_reading = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]

#counting the number of these students
pass_all_count = pass_math_reading["Student ID"].count()



# determine percentage of passing students (passing in each or both subjects / total students)
pass_math_pct = ( pass_math_count / student_count ) * 100
pass_rdg_pct = ( pass_rdg_count / student_count ) * 100
pass_all_pct = ( pass_all_count / student_count ) * 100


# setting the new dataframe titles and values
school_summary = pd.DataFrame(
        [{"Total Schools": school_count,
        "Total Students": student_count,
        "Total Budget": school_budget,
        "Average Math Score": school_math_mean,
        "Average Reading Score": school_rdg_mean,
        "% Passing Math": pass_math_pct,
        "% Passing Reading": pass_rdg_pct,
        "% Overall Passing": pass_all_pct}])

# formatting the necessary values
school_summary["Total Students"] = school_summary["Total Students"].map("{:,}".format)
school_summary["Total Budget"] = school_summary["Total Budget"].map("${:,.2f}".format)

# bonus formatting
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:.1f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:.1f}".format)
school_summary["% Passing Math"] = school_summary["% Passing Math"].map("{:.1f}".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map("{:.1f}".format)
school_summary["% Overall Passing"] = school_summary["% Overall Passing"].map("{:.1f}".format)    

school_summary.head()
```
